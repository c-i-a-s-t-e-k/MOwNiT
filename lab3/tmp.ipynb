```python
import numpy as np

def quadratic_spline(x, y):
    n = len(x)
    a = np.zeros(n-1)
    b = np.zeros(n-1)
    c = np.zeros(n-1)
    
    # obliczanie współczynników a, b i c
    for i in range(n-1):
        a[i] = y[i]
        b[i] = (y[i+1] - y[i]) / (x[i+1] - x[i])
        c[i] = ((y[i+1] - y[i]) / (x[i+1] - x[i])) / (2*(x[i+1] - x[i]))
        
    # tworzenie listy wielomianów sklejonych
    polynomials = []
    for i in range(n-1):
        poly = np.poly1d([c[i], b[i], a[i]])
        polynomials.append(poly)
    
    return polynomials

import matplotlib.pyplot as plt

# dane punkty interpolacji
x = [0, 1, 2, 3, 4]
y = [0, 2, 1, 4, 3]

# utworzenie wielomianu sklejanego
polynomials = quadratic_spline(x, y)

# rysowanie wykresu funkcji interpolowanej
plt.figure()
for i in range(len(polynomials)):
    xs = np.linspace(x[i], x[i+1], 100)
    ys = polynomials[i](xs)
    plt.plot(xs, ys, 'r-', label='interpolacja' if i==0 else None)
plt.plot(x, y, 'bo', label='punkty interpolacji')
plt.legend()
plt.show()
```
