# Badanie realizacji zadania 2
## $f(x) = (x - 1)^8$

```julia
using Plots
using PrettyTables
```

### Realizacje:

```julia
function f1(x)
    return x^8 - 8*x^7 + 28*x^6 - 56*x^5 + 70*x^4 - 56*x^3 + 28*x^2 -8*x + 1
end

function f2(x)
    return (((((((x - 8)* x + 28) * x - 56) * x + 70) * x - 56) * x + 28) * x - 8)* x + 1
end

function f3(x)
    return (x - 1)^8
end

function f4(x)
    return exp((8 * log(abs(x-1))))
end

F = [f1, f2, f3, f4]
```

### Float16 - half precision

```julia

vector::Vector{Float16} = range(0.99, stop=1.01, length=101)
Results16 = [f(x) for x in vector, f in F]

pretty_table(Results16 ; backend = Val(:html), header=["f1", "f2", "f3", "f4"])
```

#### plots

```julia
plot(vector, Results16[begin:end, 1], seriestype=:scatter)
```

```julia
plot(vector, Results16[begin:end, 2], seriestype=:scatter)
```

```julia
plot(vector, Results16[begin:end, 3], seriestype=:scatter)
```

```julia
plot(vector, Results16[begin:end, 4], seriestype=:scatter)
```
